```python
# All the imports to go here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#Reading the data from the CSV file
data1 = pd.read_csv('Iris.csv',index_col='Id')

#Copy of the data
data = data1.copy()
data.describe()
```

```python
#Label Encoding the data to put the predictions in numerical range 
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(data['Species'])
X = data.drop(['Species'],axis=1)
```

```python
#Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

#Scaling the data for better performance
scaler =StandardScaler()
Scaled_X = scaler.fit_transform(X.astype(np.float32))
Scaled_X = pd.DataFrame(Scaled_X)
Scaled_X.head()
```

```python
import random

#Making a random list for checking the errors in model 
random_seeder = []
for i in range(0,12):
    random_seeder.append(random.randint(1,100))
    
trainscore,testscore = 0,0
for seed in random_seeder:
    train_X,test_X,train_y,test_y = train_test_split(X, y, random_state=seed, test_size=0.2)
    model = LogisticRegression(penalty='none')
    model.fit(train_X,train_y)
    trainscore += model.score(train_X,train_y)
    testscore += model.score(test_X,test_y)

final_score_test = testscore/len(random_seeder)
final_score_train = trainscore/len(random_seeder)

#printing the test scores
print(final_score_train)
print(final_score_test)

model.fit(X,y)
_y= model.predict(X)
```
